# 문제1. Linear SVM vs Kernel SVM

```python
from sklearn.datasets import make_moons
import matplotlib.pyplot as plt
import numpy as np

#Toy example 생성
X, y = make_moons(n_samples=100, noise=0.1, random_state=42)

def plot_dataset(X, y, axes):
    plt.plot(X[:, 0][y==0], X[:, 1][y==0], "bs")
    plt.plot(X[:, 0][y==1], X[:, 1][y==1], "g^")
    plt.axis(axes)
    plt.grid(True, which='both')
    plt.xlabel(r"$x_1$", fontsize=20)
    plt.ylabel(r"$x_2$", fontsize=20, rotation=0)

plot_dataset(X, y, [-1.5, 2.5, -1, 1.5])
plt.show()
```

```python
# decision boundary 시각화
def plot_predictions(clf, axes):
    x0s = np.linspace(axes[0], axes[1], 100)
    x1s = np.linspace(axes[2], axes[3], 100)
    x0, x1 = np.meshgrid(x0s, x1s)
    X = np.c_[x0.ravel(), x1.ravel()]
    y_pred = clf.predict(X).reshape(x0.shape)
    y_decision = clf.decision_function(X).reshape(x0.shape)
    plt.contourf(x0, x1, y_pred, cmap=plt.cm.brg, alpha=0.2)
    plt.contourf(x0, x1, y_decision, cmap=plt.cm.brg, alpha=0.1)
```

```python
from sklearn.svm import LinearSVC

linear_svc = LinearSVC(C=1)
linear_svc.fit(X,y)
plot_predictions(linear_svc, [-1.5, 2.45, -1, 1.5])
plot_dataset(X, y, [-1.5, 2.45, -1, 1.5])
```

4. `Test set`에 대해서 `문제 2-1`에서 학습한 모델과 최종 성능에 대해서 비교하시오 (출력만 하시면 됩니다).

---

\

* 주의) 코드틀 이외의 부분에 코드를 작성하지 마세요! (해당 문제의 param_gird는 예외입니다.)
* 주의) 소스코드를 입력하고 출력 결과와 함께 캡처 후**(한 이미지에)** 보고서에 업로드하세요.
* 주의) 코드 시작과 코드 끝 라인을 삭제하지 마세요.


```python
from sklearn.model_selection import KFold

#####################문제2-2###############################

# KFold 설정
kfold = KFold(n_splits=5, shuffle=True, random_state=42)

# Accuracy가 0.8이상인 Hyperparameter을 찾을 수 있도록 수동 grid_search 수행
def grid_search(train_X, train_y, C_values, gamma_values):
    best_score = 0
    best_params = {'C': None, 'gamma': None}
    final_scores = []

    for C in C_values:
        for gamma in gamma_values:
            scores = []
            for train_index, val_index in kfold.split(train_X):
                X_train_fold, X_val_fold = train_X[train_index], train_X[val_index]
                y_train_fold, y_val_fold = train_y[train_index], train_y[val_index]
                model = SVC(kernel="rbf", gamma=gamma, C=C)
                model.fit(X_train_fold, y_train_fold)
                score = model.score(X_val_fold, y_val_fold)
                scores.append(score)
            mean_score = np.mean(scores)
            final_scores.append(scores)
            if mean_score > best_score:
                best_score = mean_score
                best_params = {'C': C, 'gamma': gamma}
                # print(f"New best score found: {best_score} with C = {C}, gamma = {gamma}")
            if best_score >= 0.8:
                break
        if best_score >= 0.8:
            break

    # 모든 점수를 하나의 배열로 통합
    all_scores = np.hstack(final_scores)
    return best_params, best_score, all_scores

# 하이퍼파라미터 값의 범위 설정
C_values = np.logspace(-3, 3, 7)
gamma_values = np.logspace(-3, 3, 7)

# 수동 그리드 서치 실행
best_params, best_score, all_scores = grid_search(train_X, train_y, C_values, gamma_values)

# 최적의 하이퍼파라미터로 최종 모델 훈련
final_svm = SVC(kernel="rbf", **best_params)
final_svm.fit(train_X, train_y)

# 교차 검증 점수의 평균과 분산 계산
mean_score = np.mean(all_scores)
variance_score = np.var(all_scores)

# 평균과 분산 출력
print(f"평균 : {mean_score}")
print(f"분산 : {variance_score}")

#########################################################

print(f"final_model 정확도 : {final_svm.score(test_X,test_y)}")
print(f"2-1 model 정확도 : {rbf_kernel_svm_clf.score(test_X,test_y)}")
```
